```python
(x, y) = (10, 25)

if x > 3 and y > 13:
    print('Both conditions are correct')
if x <= 3 or y <= 13:
    print('At least one of the conditions is false')
```

```python
x = 100;
if x > 100:
    print("A busy day")
else:
    print("A calm day")
```

```python
# if elif & else
x = 200
if x > 200:
    print("Big")
elif x > 100  and x <=200:
    print("Average")
else:
    print("Small")
```

```python
x = 200;
if x > 200:
    print("Big")
elif x > 100 and x <= 200:
    print('Average')
elif x >= 0 and x <= 100:
    print('Small')
else:
    print('Negative')
```
