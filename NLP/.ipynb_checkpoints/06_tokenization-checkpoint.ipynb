```python
# Tokenization: Converting text into smaller units called tokenization, and individual words is known as token.
# depending on use cases, token may be word, sentence, character, etc.
```

```python
# punkt_tab: a table used to figure out where sentences begin and end
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# sentence tokenization
```

```python
sentences = "Her cat's name is Luna. Her dog's name is max"
sent_tokenize(sentences)
```

```python
# word tokenization
```

```python
sentence = "Her cat's name is Luna"
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is max"
word_tokenize(sentence_2)
```
