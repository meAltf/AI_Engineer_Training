```python
# Lemmitization:
# It reduces a word to a meaningful base form while maintaining it's meaning
# It uses a predefined dictionary to keep the context of the word
# trade-off: large datasets
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# create Lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Examples
```

```python
connect_tokens = ['connecting', 'connectivity', 'connection', 'connect', 'connects', 'connected']

for t in connect_tokens:
    print(t, ': ', lemmatizer.lemmatize(t))
```

```python
learn_tokens = ['learning', 'learn', 'learns', 'learned', 'learners','learners']

for t in learn_tokens:
    print(t, ': ', lemmatizer.lemmatize(t))
```

```python
like_tokens = ['like', 'likes', 'better', 'worse']

for t in like_tokens:
    print(t, ': ', lemmatizer.lemmatize(t))
```
