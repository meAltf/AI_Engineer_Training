```python
# stemming: standardized the text
# Reducing words to their base form | It reduces the numer of unique words | reduces complexity | 
# Keeps the datasets easier to manage
```

```python
# import PorterStemmer from nltk.stem , it's a rule based algorithm to stem the words
from nltk.stem import PorterStemmer
```

```python
# create an object to store porterStemmer | necessary steps
ps = PorterStemmer()
```

```python
#examples
```

```python
connect_tokens = ['connecting', 'connectivity', 'connection', 'connect', 'connects', 'connected']

for t in connect_tokens:
    print(t, ': ', ps.stem(t))
```

```python
learn_tokens = ['learning', 'learn', 'learns', 'learned', 'learners','learners']

for t in learn_tokens:
    print(t, ': ', ps.stem(t))
```

```python
like_tokens = ['like', 'likes', 'better', 'worse']

for t in like_tokens:
    print(t, ': ', ps.stem(t))
```
