```python
# Dictionaries: Another way to storing the data
# Key: Value pair
# Needs curly braces {}
```

```python
dict = {'k1': "cat", 'k2': "dog", 'k3': "mouse", 'k4': "lion"}
dict
```

```python
dict['k2']
```

```python
dict['k4']
```

```python
# adding a new value in dictionary
dict['k5'] = "squirrel"
dict
```

```python
# Replacing existing value in dictionary
dict['k2'] = "Tiger"
dict
```

```python
# we can able to keep a list of value as well in a key
dep_worker = {'dep_1':'peter', 'dep_2':['jennifer', 'swalor', 'mohamad']}
dep_worker
```

```python
dep_worker['dep_2']
```

```python
# A new example: Another way to create a dictionary
Team = {}
Team['A'] = 'Aobert'
Team['B'] = 'Barrel'
Team['C'] = 'Culprit'
Team['D'] = 'Destroy'
```

```python
print(Team)
```

```python
print(Team['C'])
```

```python
# .get | if value exists give the same, but if not exists it gives None
print(Team.get('A'))
```

```python
print(Team.get('Rove'))
```
