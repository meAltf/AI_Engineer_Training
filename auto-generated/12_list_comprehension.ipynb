```python
numbers = [1,3,6,9,17]
numbers
```

```python
# create empty list and put twice of each elements
new_numbers = []

for n in numbers:
    new_numbers.append(n*2)
print(new_numbers)
```

```python
# same operation | 2nd method
new_numbers_2 = [ n*2 for n in numbers ]
print(new_numbers_2)
```

```python
# print values of 2 lists
for i in range(2):
    for j in range(5):
        print(i+j, end = " ")
```

```python
# same operation using list comprehension
new_list_comprehenstion_1 = [i+j for i in range(2) for j in range(5)]
new_list_comprehenstion_1
```

```python
type(new_list_comprehenstion_1)
```

```python
print(new_list_comprehenstion_1)
```

```python
# multi-dimensional list
new_list_comprehenstion_2 = [[i+j for i in range(2)] for j in range(5)]
new_list_comprehenstion_2
```

```python
type(new_list_comprehenstion_2)
```

```python
type(new_list_comprehenstion_2[1])
```

```python
type(new_list_comprehenstion_1[1])
```

```python
for i in range(1, 11):
    if(i%2 != 0):
        print(i**3, end = " ")
```

```python
# above operation using list_comprehension
[ num**3 for num in range(1,11) if num %2 != 0 ]
```

```python
# this will give syntax error | we need to give clear & cut expression
[ num**3 if num %2 != 0 for num in range(1,11) ]
```

```python
[ num**3 if num %2 != 0 else "Even" for num in range(1,11)]
```
