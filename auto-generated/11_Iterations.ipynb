```python
# While loop
x = 0
while x <= 20:
    print(x , end = " ")
    x += 2;
```

```python
def count(numbers):
    total = 0
    for n in numbers:
        if(n < 20):
            total +=1
    return total;
```

```python
# Iterate in dictionaries
prices = {
    "box_of_spaghetti" : 4,
    "lasagna"  : 5,
    "hamburger" : 2
   }
quantity = {
    "box_of_spaghetti" : 6,
    "lasagna"  : 10,
    "hamburger" : 0
    }
```

```python
money_spent = 0
for i in quantity:
    money_spent += prices[i] * quantity[i]
print(money_spent)
```

```python
money_spent = 0
for i in quantity:
    if(prices[i] >= 5):
        money_spent += prices[i] * quantity[i]
    else:
        money_spent = money_spent
print(money_spent)
```
