```python
def raise_power_2(x):
    return x ** 2;
```

```python
raise_power_2(9)
```

```python
# using lambda:
lambda x : x ** 2;
```

```python
raise_power_2_lambda = lambda x : x ** 2
```

```python
raise_power_2_lambda(9)
```
