```python
def exponentiation_exp_2(x):
    print("Raised to the power of 2:");
    result =  x**2;
    return result;
print(exponentiation_exp_2(5));
```

```python
def plus_five(x):
    return x+5;
def m_by_3(x):
    return plus_five(x) * 3;
print(m_by_3(5));
```

```python
def compare_the_two(x1, x2):
    if(x1 > x2):
        print("Greater")
    elif(x1< x2):
        print("Less")
    else:
        print("Equal")
```

```python
def subsract(a,b,c):
    result = a - b*c;
    print('Param a equals to: ', a);
    print('Param b equals to: ', b);
    print('Param c equals to: ', c);
    return result;
print(subsract(10,2,3));
```

```python
# another way to do it:
subsract(b=2, a=10, c=3)
```
